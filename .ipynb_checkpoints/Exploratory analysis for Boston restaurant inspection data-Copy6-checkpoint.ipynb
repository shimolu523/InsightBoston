# Objective:

The city of Boston performs health inspection on licensed restaurants, and published its data for 2015. The inspection result is measured with a metric referred as the violation level. Also included in the data are other identifying features such as the restaurants' names, geographic locations. The data can be matched with other restaurant features, such as Yelp rating, price range, building age, gross revenue, etc., which can be highly correlated with the violation level. We aim to build a model to predict the potential voilation levels based on these features. More specifically, the goals are:
* To perform exploratory analysis on the data, mapping out specific violation types.
* To incorporate Yelp data such as customer ratings, price range, etc., in order to model the violation levels.
* To incorporate Zillow data and add property values as a predicting feature.
* To explore Yelp comments using NLP to extract potential violation types.

# Types and levels of violation

## Violation types

We first examine the types violations, labeled as "Violation". Repeated violations might have happened for the same BusinessName, but for this step, we do not combine results for the same business.

The violation types are categorized by the first digit/letter in the violation code, according to the description:
http://www.cityofboston.gov/isd/health/violationcodes.asp

### Categorizing violation types

```python
noVio = {0:['nan']} # No violation
foodProt = {1:['M']} # Food Protection Management
food = {2:[str(x).zfill(2) for x in range(1,3)]} # Food
timeTemp = {3:[str(x).zfill(2) for x in range(3,7)]} # Time & Temperature Controls
contam = {4:[str(x).zfill(2) for x in range(7,11)]} # Food Protection From Contamination
pers = {5:[str(x).zfill(2) for x in range(11,14)]} # Personnel
eq = {6:[str(x).zfill(2) for x in range(14,27)]} # Equipment and Utensils
wat = {7:['27']} # water
sew = {8:['28']} # sewage
plumb = {9:['29','30']} # plumbing
wc = {10:['31','32']} # Toilet and Handwashing Facilities
disp = {11:['33','34']} # Refuse Disposal
anim = {12:['35']} # Insect, Rodent, Animal Control
physFac = {13:[str(x).zfill(2) for x in range(37,41)]} # Physical Facilities
othOp = {14:[str(x).zfill(2) for x in range(41,45)]} # Other Operations
highPop = {15:['45']} # Highly Susceptible Populations
unkn = {16:['36','46']}

def merge_dicts(*dict_args): # merging dictinary together, used later
    '''
    Given any number of dicts, shallow copy and merge into a new dict,
    precedence goes to key value pairs in latter dicts.
    '''
    result = {}
    for dictionary in dict_args:
        result.update(dictionary)
    return result

VioDict = merge_dicts(noVio,foodProt,food,timeTemp,contam,pers,eq,
                      wat,sew,plumb,wc,disp,anim,physFac,othOp,highPop,unkn)
VioValues = [] # all accepted values for InspDataRaw.Violation
for vioValue in VioDict.values():
    VioValues += vioValue

VioCode = {0:'Pass',
           1:'Food Protection Management',
           2:'Food',
           3:'Food Protection, Time & Temperature Controls',
           4:'Food Protection From Contamination',
           5:'Personnel',
           6:'Equipment and Utensils',
           7:'Water',
           8:'Sewage',
           9:'Plumbing',
           10:'Toilet and Handwashing Facilities',
           11:'Refuse Disposal',
           12:'Insect, Rodent, Animal Control',
           13:'Physical Facilities',
           14:'Other Operations',
           15:'Highly Susceptible Populations',
           16:'unknown code'}
```

### Cleaning data according to violation types:

```python
def categ_vioTypes(vioType, VioDict): # used to categorize violation types
    '''
    Given a string of vioType, categorize it according to the VioDict dictionary
    '''
    VioKeys = VioDict.keys()
    VioValues = VioDict.values()
    for i in range(len(VioValues)):
        if vioType in VioValues[i]:
            vioCode = VioKeys[i]
    return vioCode

# Keeping rows with only violation types matched to codes
InspDataClean = InspDataRaw.loc[InspDataRaw.Violation.astype('S32').apply(
        lambda x: x.split('-')[0] in VioValues),:]
# Replace raw data Violation codes with our category codes
Violation_cat = InspDataClean.Violation.astype('S32').apply(
    lambda x: categ_vioTypes(x.split('-')[0],VioDict))
InspDataClean.Violation = Violation_cat
```

## Levels of violation

```python
def transVioLevel(vioLevel_star): # Function to translate original violation level to numerical values
    if vioLevel_star == 'nan':
        vioLevel = 0
    else: vioLevel = len(vioLevel_star)
    return vioLevel

ViolLevel_trans = InspDataClean.ViolLevel.astype('S32').apply(transVioLevel)
InspDataClean.loc[:,'ViolLevel'] = ViolLevel_trans

ViolLevel_hist = pd.DataFrame.from_dict(Counter(InspDataClean.loc[:,'ViolLevel']), orient='index')
ViolLevel_hist.plot(kind='bar')
```

The Yelp data can be obtained by querrying its API. Methods can be found from online tutorials, such as the followingr one:
http://letstalkdata.com/2014/02/how-to-use-the-yelp-api-in-python/

We use latitude and longitude location parameters as querry keys, which is 1) unique, and 2) already available from the inspection data.

## Querry function

```python
# For validation, check Hamming distance and named id recognition

#def get_search_parameters(locLati,locLong):
def get_search_parameters(address):
    '''
    With location defined by latitude and longitude, return entire set of search parameters in Yelp API formats.
    All other parameters are set by hand below.
    See the Yelp API for more details on formats.
    '''
    #location = CityData.Address + CityData.City +
    params = {}
    params['location'] = address # string of address
    params["term"] = "food" #"restaurant"
    #params["ll"] = "{},{},{}".format(str(locLati),str(locLong),str(1000)) # last entry is the accuracy of location.
    params["radius_filter"] = "100" # radius of search circle centered at the search coordinate
    params["limit"] = "1" # limit of search results, don't change
    params["sort"] = "1"  # 1 for shortest distance from location
    
    return params

def get_results(params):
    '''
    With params in Yelp formats, querry Yelp to get desired results
    Obtain these from Yelp's manage access page
    '''
    consumer_key = "MdtGQQR4Zf7OvIU1x4jc2Q"
    consumer_secret = "VOC3Dzgd-tu-D2_GeqrNCYKyRt8"
    token = "aCf1qaYNr01MiqMx0tyzc5WdfsOmHdTL"
    token_secret = "dQuXts4pMoJEH0kgcBA24SXXgG8"
   
    session = rauth.OAuth1Session(
      consumer_key = consumer_key
      ,consumer_secret = consumer_secret
      ,access_token = token
      ,access_token_secret = token_secret)
     
    request = session.get("http://api.yelp.com/v2/search",params=params)
   
    #Transforms the JSON API response into a Python dictionary
    data = request.json()
    session.close()
   
    return data


def validation(name_yelp, name_BOS, ratioPass = 70):
    '''
    Two restaurant name strings from yelp querry and Boston City data are compared, using Hamming distance.
    A Boolean output is returned based on the benchmark score
    '''
    flag_pass = fuzz.token_set_ratio(name_yelp, name_BOS) > ratioPass
    #if flag_pass:
    #    print 'Restaurant name matched', name_BOS, 'vs', name_yelp
    #else:
    #    print 'Restaurant not found in Yelp via address:', name_BOS, 'vs', name_yelp
    return flag_pass

# Restaurant category decoding:
Categ_yelp = pd.read_json('categories.json')
print Categ_yelp.keys()
CategTable = pd.DataFrame(columns= ['Alias', 'Parent', 'Title'])
CategTable.Alias = Categ_yelp['alias']
CategTable.Parents = [x[0] if len(x)>0 else 0 for x in Categ_yelp['parents']]
CategTable.Title = Categ_yelp['title']

# Restaurant category decoding:
Categ_yelp = pd.read_json('categories.json')
CategDict = dict(zip(Categ_yelp['alias'],
                     [str(x[0]) if len(x)>0 else 0 for x in Categ_yelp['parents']]))
CategDict[0] = 0

# function to search for highest category
def findCategParent(categ, CategDict = CategDict):
    #while not {CategDict[categ] == 0 or CategDict[categ] == 'restaurants'}:
    while not CategDict[categ] == 0:
        categ = CategDict[categ]
    return categ
```

## Making querries and examine percentage of useful data

To search for restaurant from CityData in Yelp, geo coordinate were attempted as search keys. However, since the coordinate given in CityData is not the same as from Yelp, even for the same restaurant. It is found that this is NOT a systematic error shift either. This also implies using it as a validation key will not work either.

We decided to use street address for search, and restaurant name for validation. Because the registered names in the two dataset are NOT always identical, a method to quantify the differences is needed and the tolerance needs to be set as well. We decided to use Hamming distance for now, and before optimizing the comparison algorithm it's important to know roughly how much data is useful in the dataset.

## Optimization

The most basic matching (exact match) results in 223/2046 = 11% unique restaurants data useful. To improve, we investigate methods of fussy matching to improve the results. The goal is to improve to 1000 restaurants. This following website has pretty nice instructions:

http://streamhacker.com/2011/10/31/fuzzy-string-matching-python/

Result(Monday, Jan 25, 2016): Now 796/2046 = 39% of the data are found useful. So will pause on the matching algorithm and start working on the model.

We found again, there are too many categories in the data: 124 in the current data set for 796 data points in total, and twice or three times more in the total list of categories. So decided to remove this feature, evaluate the model performance, and come back to it if necessary. 

```python
# The total number of restaurants in CityData are
# len(CityData.iloc[:,0].unique())
# The CityData with one (most recent) inspection per restaurant is therefore
CityData_latest = CityData.drop_duplicates(subset = ['Address'], keep = 'last')
# print CityData_latest.iloc[:,0]
print len(CityData_latest.iloc[:,0])
# Now the useful data are 223 (exact matching)

CityData = CityData_latest
CityData.index = range(len(CityData.index))

Name_yelp = []
Rating_yelp = []
Zip_yelp = []
Categ0 = []
Categ1 = []
Categ2 = []

def MarkYelp0():
    Rating_yelp.append(0)
    Zip_yelp.append(0)
    Categ0.append(0)
    Categ1.append(0)
    Categ2.append(0)
    
for rowNo in CityData.index:
    #print rowNo
    
    address = str(CityData.Address[rowNo]) +' '+ str(CityData.City[rowNo]) +' '+ str(CityData.State[rowNo])
    params = get_search_parameters(address)
    dataYelp = get_results(params)
    name_BOS = CityData.iloc[rowNo,0]
    
    name_yelp = dataYelp['businesses'][0]['name']
    Name_yelp.append(name_yelp)
    try:
        categ_yelp = dataYelp['businesses'][0]['categories']
        rating_yelp = dataYelp['businesses'][0]['rating']
        zip_yelp = dataYelp['businesses'][0]['location']['postal_code']
        # print zip_yelp
        categ0 = categ_yelp[0][1]
        categ1 = 0
        categ2 = 0
        if len(categ_yelp) == 2:
            categ1 = categ_yelp[1][1]
        elif len(categ_yelp) == 3: 
            categ1 = categ_yelp[1][1]
            categ2 = categ_yelp[2][1]
            
        #if validation(name_yelp,name_BOS):
        if validation(name_yelp,name_BOS):
            Rating_yelp.append(rating_yelp)
            Zip_yelp.append(zip_yelp)
            Categ0.append(categ0)
            Categ1.append(categ1)
            Categ2.append(categ2)
        else:
            MarkYelp0()
            
    except KeyError:
        MarkYelp0()
        print 'missing Yelp data, restaurant not considered for model' 
```

```python
# Building most basic dataset for mysql, with simplest features:

select = [Rating_yelp[i] !=0 and Categ0[i] != 0 for i in range(len(Rating_yelp))]

CategTot = list(set(Categ0 + Categ1 + Categ2))
CategTot.remove(0)
CategTot = list(set(pd.Series(CategTot).apply(findCategParent)))

columns = ['Name','VioLevel','VioType','VioLevel_pred','VioType_pred','Rating','PropValue','Zip','locLong', 'locLati'] + CategTot

CityData_clean = pd.DataFrame(columns=columns)
#CityData_clean.Name = CityData.iloc[select,0]
CityData_clean.Name = pd.Series(Name_yelp)[select]
CityData_clean.VioLevel = CityData.ViolLevel[select]
CityData_clean.VioType = CityData.Violation[select]
CityData_clean.Rating = pd.Series(Rating_yelp)[select]
CityData_clean.PropValue = CityData.AV_TOTAL[select]
#CityData_clean.Zip = CityData.Zip[select].apply(lambda x: x.zfill(5))
CityData_clean.Zip = pd.Series(Zip_yelp)[select]
CityData_clean.locLong = CityData.LocLong.astype('float')[select]
CityData_clean.locLati = CityData.LocLati.astype('float')[select]
CityData_clean.index = range(len(CityData_clean.Name))

for i in CityData_clean.index:
    pool = pd.DataFrame([Categ0, Categ1, Categ2]).transpose()
    pool = list(pool[select].iloc[i,:].apply(findCategParent))
    for categ in CategTot:
        CityData_clean[categ][i] = int(categ in pool)

# Checking new dataset
print CityData_clean.head()
plt.figure();CityData_clean.VioLevel.hist(bins = 4)
plt.figure();CityData_clean.VioType.hist(bins = 16)
# Because the feature distributions are highly skewed, we try to transform the data

# Normalize the data
def norm_cust(data):
    return (data - data.mean())/data.std()

CityData_norm = CityData_clean.copy()

plt.figure();CityData_clean.Rating.hist(bins = 8)
CityData_norm.Rating = norm_cust(CityData_norm.Rating)
plt.figure();CityData_norm.Rating.hist(bins = 8)

plt.figure();CityData_clean.PropValue.hist(bins = 40)
CityData_norm.PropValue = norm_cust(CityData_norm.PropValue)
plt.figure();CityData_norm.PropValue.hist(bins = 40)

plt.figure();CityData_clean.locLati.hist(bins = 10)
CityData_norm.locLati = norm_cust(CityData_norm.locLati)
plt.figure();CityData_norm.locLati.hist(bins = 10)

plt.figure();CityData_clean.locLong.hist(bins = 10)
CityData_norm.locLong = norm_cust(CityData_norm.locLong)
plt.figure();CityData_norm.locLong.hist(bins = 10)
```
